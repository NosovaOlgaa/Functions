```python
'''Написать программу, которая будет на основе хранимых данных исполнять пользовательские команды'''


documents = [
{'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
{'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
{'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
'1': ['2207 876234', '11-2'],
'2': ['10006'],
'3': []
}

def is_shelf_exist(number_shelf):
    if number_shelf in directories.keys():
        return True
    return False

def new_inf(number_doc='', type_doc ='', name_doc=''):
    dict_ = {}
    dict_['number'] = number_doc
    dict_['type'] = type_doc
    dict_['name'] = name_doc
    return dict_


#Пункт 1.1 Пользователь по команде “p” может узнать владельца документа по его номеру

def identification(number):
    for number_docs in documents:
        if number == number_docs['number']:
            return number_docs['name']
              
#Пункт 1.2. Пользователь по команде “s” может по номеру документа узнать на какой полке он хранится  

def identif(numbers):
    for shelf, docs_number in directories.items():
        if numbers in docs_number:
            return shelf

#Пункт 1.3. Пользователь по команде “l” может увидеть полную информацию по всем документам

def all_information():
    for all_inf in documents:
        print(f'№: {all_inf["number"]}, тип {all_inf["type"]}, владелец: {all_inf["name"]}, полка хранения: {identif(all_inf["number"])} ')  

#Пункт 1.4. Пользователь по команде “as” может добавить новую полку

def add_new_shelf(shelf_num):
    if shelf_num in directories.keys():
        print(f'Такая полка уже существует.')
    else:
        directories.setdefault(shelf_num, [])
        print(f'Добавлена полка № {shelf_num}')
        print(f'Текущий перечень полок: {",".join(directories.keys())}')

#Пункт 1.5. Пользователь по команде “ds” может удалить существующую полку из данных (только если она пустая)

def del_shelf(shelf_):
    if directories.get(shelf_):
        print(f'На полке есть документы, удалите их перед удалением полки.')
    elif shelf_ not in directories.keys():
        print(f'Такой полки не существует.')
    else:
        directories.pop(shelf_, [])
        print(f'Удалена полка № {shelf_}')
        print(f'Текущий перечень полок: {",".join(directories.keys())}')

#Пункт 2.1. Пользователь по команде “ad” может добавить новый документ в данные

def add_new_doc(shelf_number = '', number_doc = '', type_doc = '', name_doc = ''):
    if not is_shelf_exist(shelf_number):
        return f'Такой полки не существует. Добавьте полку командой as.'
        return
    new_doc = new_inf(
            number_doc = input('Введите номер документа: '),
            type_doc = input('Введите тип документа: '),
            name_doc = input('Введите владельца документа: ')
    )
    documents.append(new_doc)
    directories[shelf_number].append(new_doc['number'])
    print('Документ добавлен. Текущий список документов:')
    all_information()


#Пункт 2.2. Пользователь по команде “d” может удалить документ из данных

#удаление данных из documents
def del_doc_from_documents(doc_number = ''):
    for index, doc in enumerate(documents):
        if doc_number == doc['number']:
            return documents.pop(index)
        
#удаление данных из directories
def del_shelf_from_directories(doc_number = ''):
    for element in directories.values():
         if doc_number in element:
            return element.pop(element.index(doc_number))
        
def del_docs(number_doc = ''):
    if del_doc_from_documents(number_doc) and del_shelf_from_directories(number_doc):
        print('Документ удален. Текущий список документов:')
    else:
        print('Документ не найден в базе. Текущий список документов:')
        

#Пункт 2.3. Пользователь по команде “m” может переместить документ с полки на полку

def move_document_shelf_to_shelf(number_doc = '', shelf_num = ''):
    doc = del_shelf_from_directories(number_doc)
    if shelf_num not in directories.keys():
        print(f'Такой полки не существует.Текущий перечень полок: {",".join(directories.keys())}' )
        return
    if not doc:
        print('Документ не найден в базе.Текущий список документов:')
        all_information()
        return    
    if number_doc in directories[shelf_num]:
        print('Документ уже находится на указанной полке')
        return
    directories[shelf_num].append(doc)
    print('Документ перемещен')
    all_information()
```

```python
def main():
    while True:
        command = input('Введите команду ')
        if command == 'p':
            doc_owner = identification(input('Введите номер документа '))
            if doc_owner:
                print(f'Владелец документа: {doc_owner}')
            else:
                print('Документ не найден в базе')
        elif command == 'q':
            return
        elif command == 's':
            num_shelf = identif(input('Введите номер документа '))
            if num_shelf:
                print(f'Документ хранится на полке: {num_shelf}')
            else:
                print('Документ не найден в базе') 
        elif command == 'l':
            all_information()
        elif command == 'as':
            add_new_shelf(input('Введите номер полки: ')) 
        elif command == 'ds':
            del_shelf(input('Введите номер полки: ')) 
        elif command == 'ad':
            add_new_doc(shelf_number = input('Введите номер полки '))
        elif command == 'd':
            del_docs(input('Введите номер документа: '))
            all_information()
        elif command == 'm':
            move_document_shelf_to_shelf(
                input('Введите номер документа: '),
                input('Введите номер полки: ')
            )    
```

```python
main()
```
